## Encoding

```python
import alphapy as ap
from alphapy.nlp import encode_price
import glob
import pandas as pd
```

```python
group_dict = {'1D'  : ('boy', 'eoy'),
              '1H'  : ('bom', 'eom'),
              '5min' : ('bod', 'eod'),
              '1min' : ('bod', 'eod')}
group_dict
```

```python
subject = 'crypto'
fractal = '1min'
period = 20
```

```python
root_path = '/Users/markconway/Projects/alphapy-root/alphapy-markets'
input_str = 'data'
data_dir = '/'.join([root_path, input_str])
output_str = 'nlp-data'
nlp_dir = '/'.join([root_path, output_str])
```

```python
input_dir = '/'.join([data_dir, subject, fractal])
output_dir = '/'.join([nlp_dir, subject, fractal])
input_dir, output_dir
```

```python
search_spec = '/'.join([input_dir, '*.csv'])
file_specs = glob.glob(search_spec)
len(file_specs)
```

```python
output_files = [s.replace('/'.join([input_str, subject]), '/'.join([output_str, subject])) for s in file_specs]
len(output_files)
```

```python
test_encoding = False
if test_encoding:
    df_test = pd.read_csv(file_specs[0])
    print(df_test)
    encode_price(df_test, p=period, intraday=False)
```

```python
def write_data(input_files, output_files, fractal, group_dict, dt_col):
    intraday = group_dict[fractal]
    for input_file, output_file in zip(input_files, output_files):
        file_name = input_file.split('/')[-1]
        symbol = file_name.split('_')[0]
        print("Processing %s" % symbol)
        df_in = pd.read_csv(input_file)
        records = []
        if fractal == '1D':
            group_col = 'year'
            df_in[group_col] = df_in[dt_col].str[:4]
        elif fractal == '1H':
            group_col = 'year_month'
            df_in[group_col] = df_in[dt_col].str[:7]
        else:
            group_col = 'date'
            df_in[group_col] = df_in[dt_col].str[:10]
        for group_name, df_group in df_in.groupby(group_col):
            text = encode_price(df_group, p=period, intraday=intraday)
            record = [symbol, subject, fractal, period, group_name, text]
            records.append(record)
        df_out = pd.DataFrame(records, columns=['symbol', 'subject', 'fractal', 'period', 'date', 'text'])
        print("Writing %s" % output_file)
        df_out.to_csv(output_file, index=False)
```

```python
dt_col = 'datetime'
write_data(file_specs, output_files, fractal, group_dict, dt_col)
```
